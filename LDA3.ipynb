```python
import pandas as pd
import pyLDAvis
import pyLDAvis.gensim_models
import os
import gensim
import nltk
from gensim import corpora
from gensim.models.ldamodel import LdaModel
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('1.csv')
```

```python
new_column_names = {old_name: str(i + 1) for i, old_name in enumerate(df.columns)}

# Apply the new column names to the DataFrame
df_renamed = df.rename(columns=new_column_names)
```

```python
columns_to_keep_1 = [
    "49"
]
df3 = df_renamed[columns_to_keep_1]
```

```python
df3 = df3.dropna()
```

```python
# Load standard stop words
stop_words = set(stopwords.words('english'))

# Add custom stop words
custom_stop_words = [
    'seems','ca','wo','use','want', 'and', 'the', 'i', 'changed','changes', 
    'stay', 'one', 'notice','seen','made','since', 'said', 'say','to','is','was','for','in','of','The','but','at','a','an','as','had','have','on','from','you','would','will','not','us','we','get','could'  # Ensure not to duplicate words
]
stop_words.update(custom_stop_words)

# Function to generate tokens
def generate_tokens(text):
    # Tokenize and filter out stop words and any other potential noise
    return [word for word in word_tokenize(text.lower()) if word not in stop_words and word.isalpha()]

# Apply the tokenization function to the 'Review' column
df3['tokens'] = df3['49'].apply(generate_tokens)
```

```python
def create_dictionary(words):
    return corpora.Dictionary(words)
id2word=create_dictionary(df3['tokens'])
```

```python
def create_document_matrix(tokens,id2word):
    corpus = []
    for text in tokens:
        corpus.append(id2word.doc2bow(text))
    return corpus
#passing the dataframe column having tokens and dictionary
corpus=create_document_matrix(df3['tokens'],id2word)
print(df3['tokens'][0])
print(corpus[0])
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                            id2word=id2word,
                                            num_topics=10,
                                            random_state=100,
                                             ) #building up LDA momdel
```

```python
#applying LDA model
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word, mds="mmds", R=30)
vis 
```

```python
# analyse csr report topic            
# Standard NLTK stop words
stop_words = set(stopwords.words('english'))

# Add custom stop words
custom_stop_words = [
    'seems','ca','wo','use','want', 'and', 'the', 'i', 'changed','changes', 
    'stay', 'one', 'notice','seen','made','since', 'said', 'say','to','is','was','for','in','of','The','but','at','a','an','as','had','have','on','from','you','would','will','not','us','we','get','could']
stop_words.update(custom_stop_words)

# Apply LDA
num_topics = 10  
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=id2word, passes=15)

# Print the topics
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```
